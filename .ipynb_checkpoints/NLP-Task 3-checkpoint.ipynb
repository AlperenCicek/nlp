```python
#Generating n-grams of string
def extract_ngrams(data, num):
    n_grams = ngrams(data, num)
    return [' '.join(grams) for grams in n_grams]
```

```python
#Counting which word how many there is in the string
def countingWord(data, index):
    count = 0
    for j in range(len(data)):
        if data[index] == data[j]:
            count = count + 1
    return data[index] + " " + str(count)
```

```python
#Creating Unique Word List
def creatingUniqueWordList(data):
    word_list = []
    for i in range(len(data)):
        count = 0
        isItThereOnList = False
        for j in range(len(data)):
            if data[i] == data[j]:
                count = count + 1
        for k in range(len(word_list)):
            if word_list[k][0] == data[i]:
                isItThereOnList = True
        if isItThereOnList != True:
            word_list.append([data[i], count])
    return word_list
```

```python
#Getting Words More Than One
def wordsMoreThanOne(data):
    words_more_than_one = []
    for i in range(len(data)):
        if data[i][1] > 2:
            words_more_than_one.append(data[i])
    return words_more_than_one
```

```python
import nltk
nltk.download('averaged_perceptron_tagger')
def gettingAllPostagCount(data):
    postTags = []
    tagWord = nltk.pos_tag(data)
    for i in range(len(tagWord)):
        postTags.append(tagWord[i][1])
    postTags = dict((i, postTags.count(i)) for i in postTags)
    return postTags
```

```python
def gettingPostagCountFromTwoText(data1, data2, element):
    postTagsData1 = []
    postTagsData2 = []
    tagWord = nltk.pos_tag(data1)
    for i in range(len(tagWord)):
        postTagsData1.append(tagWord[i][1])
    tagWord = nltk.pos_tag(data2)
    for i in range(len(tagWord)):
        postTagsData2.append(tagWord[i][1])
    return element.upper() + " --> text1: " + str(postTagsData1.count(element)) + " text2: " + str(postTagsData2.count(element))
```
