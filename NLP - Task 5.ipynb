```python
import nltk
import pandas as pd
import numpy as np
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.decomposition import TruncatedSVD
from nltk.util import ngrams
from nltk.tokenize import RegexpTokenizer
import gensim

bodytext = open(r'.\text.txt', 'r', encoding='utf-8')
body = bodytext.readlines()
text = [""]
for sentence in body:
    text = [text[0] + str(sentence) + " "]
text
```

```python
#Generating n-grams of string
def extract_ngrams(data, num):
    for sentence in data:
        n_grams = ngrams(RegexpTokenizer(r'\w+').tokenize(sentence), num)
        return [ ' '.join(grams) for grams in n_grams]
```

```python
#Counting which word how many there is in the string
def countingWord(data, word):
    count = 0
    for i in range(len(data)):
        if data[i] == word:
            count = count + 1
    return word + " " + str(count)
```

```python
#Creating Unique Word List
def creatingUniqueWordList(data):
    word_list = []
    for i in range(len(data)):
        count = 0
        isItThereOnList = False
        for j in range(len(data)):
            if data[i] == data[j]:
                count = count + 1
        for k in range(len(word_list)):
            if word_list[k][0] == data[i]:
                isItThereOnList = True
        if isItThereOnList != True:
            word_list.append([data[i], count])
    return word_list
```

```python
#Getting Words More Than One
def wordsMoreThan(data, num):
    words_more_than = []
    for i in range(len(data)):
        if data[i][1] >= num:
            words_more_than.append(data[i])
    return words_more_than
```

```python
def gettingTokens(data):
    word_tokens_without_punctuation = RegexpTokenizer(r'\w+')
    word_tokens = []
    for sentence in data:
        word_tokens.append(word_tokens_without_punctuation.tokenize(sentence))
    return word_tokens
```

```python
def gettingFrequency(data):
    for i in range(len(data)):
        data[i][1] = data[i][1] / len(data)
    return data
```

```python
def eleminatingStopWords(text):
    from nltk.corpus import stopwords
    stop_words = set(stopwords.words('english')) 
    data = []
    for i in text:
        if i not in stop_words:
            data.append(i)
    return data
```

```python
from nltk.tokenize import word_tokenize, sent_tokenize
```

```python
file_docs = body
file_docs
```

```python
gen_docs = [[w.lower() for w in word_tokenize(text)] 
            for text in file_docs]
gen_docs
```

```python
dictionary = gensim.corpora.Dictionary(gen_docs)
dictionary.token2id
```

```python
corpus = [dictionary.doc2bow(gen_doc) for gen_doc in gen_docs]
corpus
```

```python
tf_idf = gensim.models.TfidfModel(corpus)
```

```python
sims = gensim.similarities.Similarity('./',tf_idf[corpus],
                                        num_features=len(dictionary))
```
