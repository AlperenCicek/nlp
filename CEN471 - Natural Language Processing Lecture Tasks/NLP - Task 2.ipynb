```python
#Generating n-grams of string
def extract_ngrams(data, num):
    n_grams = ngrams(RegexpTokenizer(r'\w+').tokenize(data), num)
    return [ ' '.join(grams) for grams in n_grams]
```

```python
#Counting which word how many there is in the string
def countingWord(data, index):
    count = 0
    for j in range(len(data)):
        if data[index] == data[j]:
            count = count + 1
    return data[index] + " " + str(count)
```

```python
#Creating Unique Word List
def creatingUniqueWordList(data):
    word_list = []
    for i in range(len(data)):
        count = 0
        isItThereOnList = False
        for j in range(len(data)):
            if data[i] == data[j]:
                count = count + 1
        for k in range(len(word_list)):
            if word_list[k][0] == data[i]:
                isItThereOnList = True
        if isItThereOnList != True:
            word_list.append([data[i], count])
    return word_list
```

```python
#Getting Words More Than One
def wordsMoreThanOne(data):
    words_more_than_one = []
    for i in range(len(data)):
        if data[i][1] != 1:
            words_more_than_one.append(data[i])
    return words_more_than_one
```
