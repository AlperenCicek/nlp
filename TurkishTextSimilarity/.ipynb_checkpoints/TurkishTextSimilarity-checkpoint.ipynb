```python
#Tokenization işlemi verilen metindeki  herbir kerlimenin ayrılma sürecine denir bu fonksiyon sayesinde her kelimeyi tane tane alabilmekteyiz ve noktalama işaretlerinden de kurtulabilmekteyiz.
#Örnek: "Ben 10 yaşındayım." -> ["Ben", "10", "yaşındayım"]
def gettingTokens(data):
    word_tokens_without_punctuation = RegexpTokenizer(r'\w+')
    word_tokens = []
    for sentence in data:
        word_tokens.append(word_tokens_without_punctuation.tokenize(sentence))
    return word_tokens
```

```python
#Doğal dil işleme metotlarını yapmadan önce konudaki kilit kelimelerin analizi önemlidir. Bundan dolayı gereksiz artık olarak nitelendirilen bağlaç gibi kelimeler içerikten temizlenmelidir. Bu fonksiyon ile "gibi, ve, ile" gibi kelimeler filtrenir.
#Örnek: "ile", "gibi", "ancak"
def eleminatingStopWords(data):
    stopwords = open(r'.\stop_words_turkish.txt', 'r', encoding='utf-8')
    bodyStopWord = stopwords.readlines()
    stopwordsArray = []
    for sentence in bodyStopWord:
        stopwordsArray.append(str(sentence))
    stopwordsArray = gettingTokens(stopwordsArray)
    eleminatedStopWordArray = []
    
    for word in data:
        isItAvaliable = False
        for stopword in stopwordsArray:
            if [word] == stopword:
                isItAvaliable = True
        if isItAvaliable == False:
            eleminatedStopWordArray.append(word)
    return eleminatedStopWordArray
```

```python
#Kilit kelimelerin bulunması sonucunda eklerden arındırılması da gerekir böylece kelime köklerini karşılaştırmak daha verimli  olmaktadır.
#Örnek: geliyorum -> gelmek, dinliyorlar -> dinlemek
def lemmatizingText(data):
    lemmatizer = MorphAnalyzer()
    lemmatizedArray = []
    for word in data:
        lemmatizedArray.append(lemmatizer.lemmatize(word)[0])
    return lemmatizedArray
```

StopWord'lerin atılması sonrası: ["Arkadaşları"], ["arabası"], ["olsun"], ["istiyordu"]

Lemmatization sonrası: ["Arkadaş"], ["araba"], ["olmak"], ["istemek"]

DAHA SONRA KELİME KÖKLERİNİN BENZERLİK İLİŞKİSİNE BAKILIR

```python
def levenshteinDistanceDP(token1, token2):
    distances = np.zeros((len(token1) + 1, len(token2) + 1))

    for t1 in range(len(token1) + 1):
        distances[t1][0] = t1

    for t2 in range(len(token2) + 1):
        distances[0][t2] = t2
        
    a = 0
    b = 0
    c = 0
    
    for t1 in range(1, len(token1) + 1):
        for t2 in range(1, len(token2) + 1):
            if (token1[t1-1] == token2[t2-1]):
                distances[t1][t2] = distances[t1 - 1][t2 - 1]
            else:
                a = distances[t1][t2 - 1]
                b = distances[t1 - 1][t2]
                c = distances[t1 - 1][t2 - 1]
                
                if (a <= b and a <= c):
                    distances[t1][t2] = a + 1
                elif (b <= a and b <= c):
                    distances[t1][t2] = b + 1
                else:
                    distances[t1][t2] = c + 1
    return distances[len(token1)][len(token2)]
```

Levenshtein Distance uzaklık tabanlı benzerlik algoritmasıdır. Bu algoritmanın çalışma prensibi harflerin sırayla karşılaştırıp farkının alınmasıdır.

Örnek: OZAN - OKAN arasındaki benzerlik ilk harf benzerdir 0 gelir. iİkinci harf farklıdır 1 gelir. Üçüncü harf benzerdir 0 gelir. Dördüncü harf benzerdir 0 gelir. Böylece bu benzerliğin skoru 1dir. Yani 1 harf farklıdır. Skor ne kadar az olursa kelimeler o kadar benzerdir

```python
# Bu fonksiyon ile her iki text'in kelimelerini kıyaslayarak en benzer olanları elimizde tutuyoruz ve skorlarını topluyoruz. Kelime sayısına bölerek de yüzdelik bir başarı elde edebiliyoruz.
def findSimilarity(data1, data2):
    similarityWeight = 0
    sameWords = []
    wordCounter = 0
    for word1 in data1:
        wordCounter = wordCounter + 1
        minSimilarityWeight = 10
        for word2 in data2:
            #print(word1, word2)
            score = levenshteinDistanceDP(word1, word2)
            #print(score)
            if score <= minSimilarityWeight:
                minSimilarityWeight = score
        if minSimilarityWeight == 0:
            minSimilarityWeight = 1
            sameWords.append(word1)
        else:
            minSimilarityWeight = 0
        similarityWeight = similarityWeight + minSimilarityWeight
    totalScore = similarityWeight / wordCounter
    print("Her iki metinde de geçen ortak kelime kökleri : \n", sameWords)
    print("*************************************************************")
    print("Skor (0 : Tamamen farklı / 1 : Birebir aynısı) : ", totalScore)
```
